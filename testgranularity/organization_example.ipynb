#Instances Example

In this example we created several different instances and each of them has unit tests corresponding to only a single operator of our calculator.

```
import unittest

class Calculator:
    
    def add(a, b):
        return a + b

    def sub(a, b):
        return a - b
  
    def mul(a, b):
        return a * b

    def div(a, b):
        if b != 0:
            return a / b
```

```
class TestCalcAddFunc(unittest.TestCase):

    def test_add_units(self):
        self.assertEqual(Calculator.add(2, 4), 6)
        self.assertEqual(Calculator.add(6, 3), 9)

    def test_add_dozens(self):
        self.assertEqual(Calculator.add(10, 10), 20)
        self.assertEqual(Calculator.add(10, 55), 65)
```

# SetUp & TearDown Example

First step: import Python's Unit Testing Framework.

```
import unittest
```

Let's start with a simple class called Person. We can get the name from a person and a self-introduction.

```
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def getName(self):
        return f'My name is {self.name}'

    def getIntro(self):
        return f"{self.getName()} and I'm {self.age}"
```

We need to test `getName()` and `getIntro()` are working correctly.

```
class PeopleTestCase(unittest.TestCase):
    
    def test_name(self):
        person = Person("John", 36)
        self.assertEqual(person.getName(), 'My name is John')

    def test_intro(self):
        person = Person("John", 36)
        self.assertEqual(person.getIntro(), "My name is John and I'm 36")

unittest.main(argv=[''], verbosity=2, exit=False)
```

Notice we are creating the same object (person) manually in every test case. An alternative could be using the `setUp` function to avoid duplicating code.

```
class PeopleTestCase(unittest.TestCase):

    # Runs before any test
    def setUp(self):
        self.person = Person("John", 36)
    
    def test_name(self):
        self.assertEqual(self.person.getName(), 'My name is John')

    def test_intro(self):
        self.assertEqual(self.person.getIntro(), "My name is John and I'm 36")

    # Create your own tearDown() here...

unittest.main(argv=[''], verbosity=2, exit=False)
```

Unlike `setUp`, `tearDown` will always be run after each test case.
