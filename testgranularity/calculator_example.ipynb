# Calculator Test Case

Import the Python's Unit Testing framework.

```
import unittest
```

Calculator is a simple class with only four methods to be tested.

```
class Calculator:
    
    # Addition
    def add(a, b):
        return a + b

    # Subtraction
    def sub(a, b):
        return a - b
  
    # Multiplication
    def mul(a, b):
        return a * b

    # Division
    def div(a, b):
        if b != 0:
            return a / b
```

A test case is created by subclassing `unittest.TestCase`.

```
class TestCalculator(unittest.TestCase):
  
    def test_add(self):
        '''Test case function for addition'''
        result = Calculator.add(4, 7)
        expected = 11
        self.assertEqual(result, expected)

    def test_sub(self):
        '''Test case function for subtraction'''
        result = Calculator.sub(10, 5)
        expected = 6
        self.assertEqual(result, expected)

    @unittest.skip('Some reason')
    def test_mul(self):
        '''Test case function for multiplication'''
        result = Calculator.mul(3, 7)
        expected = 21
        self.assertEqual(result, expected)

    def test_div(self):
        '''Test case function for division'''
        result = Calculator.div(10, 2)
        expected = 5
        self.assertEqual(result, expected)
```

We created 4 unit tests, each of it is checking a method of the Calculator class. These checks are being done through calls to `Assertions`, in this case the `assertEqual` function. Note that, flagging the method `test_mul` with `@unittest.skip('your_reason')` will skip the test for that method. For mastering `skipping test`, see the [doc](https://docs.python.org/3/library/unittest.html#skipping-tests-and-expected-failures).

To simply run the tests call:


```
unittest.main(argv=[''], verbosity=1, exit=False)
```

By checking the very first line of the output you will see: 

- **.** meaning the test has passed;
- **F** meaning the test has failed;
- **s** meaning the test was skipped.

This is due to the default verbosity level of 1. Try setting `verbosity=2` for a more detailed output.
