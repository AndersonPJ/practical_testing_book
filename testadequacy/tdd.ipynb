## TDD

Test-Driven Develoment has become a very popular pratice among developers. But, what is TDD? The concept is surprisingly simple, before writting the implementation code, write the test to prove whether your implementation works or not. The conventional approach makes us write the feature and then the test, with TDD approach the process is the opposite. 

### Benefits

We can quote some benefits of TDD:

Writting the tests first, you will be able to understand the software requirement much better; 

Increase the code quality;

TDD forces you to write classes with single responsability (A SOLIDâ€™s principle);

TDD approach reinforces the fail fast mindset, keep aligned with the business;

Don't worry if the test failed, we completed the red stage of TDD cycle.
